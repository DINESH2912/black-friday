```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer

# Assuming 'data' is your DataFrame containing the necessary columns
# Assuming 'City_Category' is a feature column and 'Purchase' is the target column

# Separate features and target
features = data['City_Category']
target = data['Purchase']

# Create a linear regression model
model = LinearRegression()

# Fit the model on available data
model.fit(features, target)

# Identify rows with missing target values
missing_target_rows = target.isnull()

# Extract features for rows with missing target values
features_missing_target = features[missing_target_rows]

# Predict missing target values
predicted_targets = model.predict(features_missing_target)

# Fill in the missing target values in the original DataFrame
data.loc[missing_target_rows, 'Purchase'] = predicted_targets
```
